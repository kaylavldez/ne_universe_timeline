```python
import json
import requests
import pandas as pd

with open("token.json") as f:
    token = json.load(f)["access_token"]
    
headers = {
    "Authorization": f"Bearer {token}"
}
     
# Function to search for the artist
def search_artist_id(artist_name):
    search_url = "https://api.spotify.com/v1/search"
    params = {
        "q": artist_name,
        "type": "artist",
        "limit": 1
    }
    
    res = requests.get(search_url, headers = headers, params = params)
    artist_data = res.json()
    artist_id = artist_data["artists"]["items"][0]["id"]
    
    return artist_id
        
```

<h1>New Edition</h1>

```python
# Get artist id
artist_id = search_artist_id("New Edition")

# Get the albums
albums_url = f"https://api.spotify.com/v1/artists/{artist_id}/albums"
params = {
    "limit": 20,
    "include_groups": "album, single"
}
res = requests.get(albums_url, headers = headers, params = params)
albums = res.json()["items"]

# Convert to DataFrame
ne_df = pd.DataFrame([
    {
        "album": album["name"],
        "releaseDate": album["release_date"],
        "totalTracks": album["total_tracks"],
        "spotifyURL": album["external_urls"]["spotify"]
    }
    for album in albums
])

# Fix data for future manipulation
ne_df.loc[10, "releaseDate"] = "1983-01-01"
ne_df.loc[11, "releaseDate"] = "1983-01-01"

ne_df
```

<h1>LSG</h1>

```python
# Get artist id
artist_id = search_artist_id("LSG")

# Get the albums
albums_url = f"https://api.spotify.com/v1/artists/{artist_id}/albums"
params = {
    "limit": 20,
    "include_groups": "album, single"
}
res = requests.get(albums_url, headers = headers, params = params)
albums = res.json()["items"]

# Convert to DataFrame
lsg_df = pd.DataFrame([
    {
        "album": album["name"],
        "releaseDate": album["release_date"],
        "totalTracks": album["total_tracks"],
        "spotifyURL": album["external_urls"]["spotify"]
    }
    for album in albums
])

lsg_df
```

```python
ne_df["artist"] = "New Edition"
bb_df["artist"] = "Bobby Brown"
rt_df["artist"] = "Ralph Tresvant"
jg_df["artist"] = "Johnny Gill"
bbd_df["artist"] = "Bell Biv DeVoe"
lsg_df["artist"] = "LSG"

df = pd.concat([ne_df, bb_df, rt_df, jg_df, bbd_df, lsg_df], ignore_index = True)
df["releaseDate"] = pd.to_datetime(df["releaseDate"])
df = df.sort_values("releaseDate")

df = df[["artist"] + [col for col in df.columns if col != "artist"]]
df = df.drop_duplicates("album")
df = df.reset_index(drop = True)

# Cutting off all music after the "25 Strong: The BET Silver Anniversary Special"
df = df[df["releaseDate"] <= pd.to_datetime("2005-10-26")]
df
```

```python
# Save data to a file
df.to_json('albums.json', orient='records', indent=2)
```
